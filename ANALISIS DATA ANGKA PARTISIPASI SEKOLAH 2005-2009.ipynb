# ANALISIS DATA
Analisis data Angka Partisipasi sekolah tahun 2005-2009 di Indonesia

```python
import pandas as pd

aps = pd.read_csv('processed-angka-partisipasi-sekolah-13-15-2005-2009.csv', delimiter = ',', encoding = "ISO-8859-1")
aps
```

# DESKRIPSI DATA

Angka partisipasi sekolah tiap kabupaten dan kota di Indonesia pada tahun 2005-2009. Tiap kabupaten dan kota juga digolongkan sesuai provinsinya. 

# STATISTIK DATA

```python
s = aps.angka_partisipasi_sekolah.sum()
mi = aps.angka_partisipasi_sekolah.min()
ma = aps.angka_partisipasi_sekolah.max()

print("Jumlah:",s,"Angka Partisipasi minimal:",mi,"Angka Partisipasi maksimal:",ma)
```

```python
me = aps.angka_partisipasi_sekolah.mean()
md = aps.angka_partisipasi_sekolah.median()
mo = aps.angka_partisipasi_sekolah.mode()
```

```python
print("MEAN:")
print(me)
print("MEDIAN:")
print(md)
print("MODUS:")
print(mo)
```

# VISUALISASI SEDERHANA
Visualisasi perkembangan angka partisipasi sekolah rata-rata pertahun untuk seluruh Indonesia.

```python
grp=aps.groupby('tahun')["angka_partisipasi_sekolah"].mean()
aps.groupby('tahun')["angka_partisipasi_sekolah"].mean()
```

```python
ax = aps.groupby('tahun').mean()['angka_partisipasi_sekolah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Angka Partisipasi Sekolah", fontsize=12)
```
